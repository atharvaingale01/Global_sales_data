## Data Transformation

```python
df_final['sales_id'] = df_final.index
```

```python
df_final.head()
```

## Shipping Table Dim

```python
shipping_dim = df_final[["Order ID","Order Date","Ship Date"]].reset_index(drop=True)
```

```python
shipping_dim["Ship Date"] = pd.to_datetime(shipping_dim["Ship Date"],format="%m/%d/%Y")
shipping_dim["Order Date"] = pd.to_datetime(shipping_dim["Order Date"],format="%m/%d/%Y")
```

```python
shipping_dim.head()
```

```python
shipping_dim['Order_hour'] = shipping_dim['Order Date'].dt.hour
shipping_dim['Order_day'] = shipping_dim['Order Date'].dt.day
shipping_dim['Order_month'] = shipping_dim['Order Date'].dt.month
shipping_dim['Order_year'] = shipping_dim['Order Date'].dt.year
shipping_dim['Order_weekday'] = shipping_dim['Order Date'].dt.weekday
```

```python
shipping_dim['Ship_hour'] = shipping_dim['Ship Date'].dt.hour
shipping_dim['Ship_day'] = shipping_dim['Ship Date'].dt.day
shipping_dim['Ship_month'] = shipping_dim['Ship Date'].dt.month
shipping_dim['Ship_year'] = shipping_dim['Ship Date'].dt.year
shipping_dim['Ship_weekday'] = shipping_dim['Ship Date'].dt.weekday
```

```python
shipping_dim = shipping_dim[["Order ID","Order Date","Order_hour","Order_day","Order_month","Order_year","Order_weekday",
                             "Ship Date","Ship_hour","Ship_day","Ship_month","Ship_year","Ship_weekday"]]
```

```python
shipping_dim.head()
```

## Sales channel dim

## Order Priority Dim

```python
Order_Priority_dim_unique = df_final[["Order Priority"]].drop_duplicates().reset_index(drop=True)
Order_Priority_dim_unique["Order_Priority_id"] = Order_Priority_dim_unique.index
```

```python
Order_Priority_dim_unique.head()
```

```python
order_priority_name = {
    "H": "High Priority",
    "L": "Low Priority",
    "M": "Medium Priority",
    "C": "Critical Priority",
}

Order_Priority_dim['Order Priority'] = Order_Priority_dim['Order Priority'].map(order_priority_name)
```

```python
Order_Priority_dim_unique
```

# Fact Table

```python
Fact_table = df_final.merge(shipping_dim,left_on='Order ID', right_on='Order ID') \
            .merge(Region_dim_unique,left_on='Region', right_on='Region') \
            .merge(Country_dim_unique,left_on='Country', right_on='Country') \
            .merge(item_type_dim_uni,left_on='Item Type', right_on='Item Type') \
            .merge(Sales_Channel_dim_unique,left_on='Sales Channel', right_on='Sales Channel') \
            .merge(Order_Priority_dim_unique,left_on='Order Priority', right_on='Order Priority')
```
